```python
# coding: utf-8
"""
ネットーワークXを用いてグラフクラスタリングを行う
"""
import networkx as nx
import csv
import community

# csvファイルの読み込み
def readcsv(path):
    f = open(path, "rb")
    dataReader = csv.reader(f)
    arr = [row for row in dataReader]
    return arr

# csvファイルの書き込み
def writecsv(arr, path):
    f = open(path, "ab")
    dataWriter = csv.writer(f)
    dataWriter.writerows(arr)
    f.close()
```

```python
list_edge = readcsv("./files/text_for_cytoscape_rev.csv")
list_edge = [tuple(row) for row in list_edge]
```

```python
G=nx.Graph()
G.add_edges_from(list_edge)
```

```python
DG=nx.Graph()
DG.add_edges_from(list_edge)
```

```python
print G.number_of_nodes()
print G.number_of_edges()
print DG.number_of_nodes()
print DG.number_of_edges()
```

```python
partition = community.best_partition(G)
```

```python
size = float(len(set(partition.values())))
```

```python
dict_cluster = {}
for row in partition.items():
    if dict_cluster.has_key(row[1]):
        dict_cluster[row[1]].append(row[0])
    else:
        dict_cluster[row[1]] = [row[0]]
```
